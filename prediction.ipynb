```python
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import RobustScaler
from sklearn.metrics import r2_score, mean_squared_error
```

```python
normDNAmeth = pd.read_csv('normDNAmeth.csv')
normCNA = pd.read_csv('normCNA.csv')
```

```python
print(type(normDNAmeth))
```

```python
normDNAmeth.shape
```

```python
input_neurons = normDNAmeth.shape[1]
output_neurons = normCNA.shape[1]
print('No of i/p neurons = ',input_neurons,' and no of o/p neurons = ',output_neurons)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(normDNAmeth, normCNA, test_size=0.2, random_state=7)
```

```python
print('X_train shape:', X_train.shape, 'X_test shape:', X_test.shape, 'Y_train shape:', Y_train.shape, 'Y_test shape:', Y_test.shape)
```

```python
X_train
```

```python
rc = RobustScaler()
X_train_norm = rc.fit_transform(X_train)
X_test_norm = rc.transform(X_test)
```

```python
print(type(X_train_norm))
```

```python
X_train_norm
```

```python
print(type(X_test_norm))
```

```python
X_test_norm
```

#### Setting up ANN

```python
ann = tf.keras.models.Sequential()
```

```python
ann.add(tf.keras.layers.Dense(units=500, activation="relu"))  # input layer
ann.add(tf.keras.layers.Dense(units=200, activation="relu"))  # 1st hidden layer
ann.add(tf.keras.layers.Dense(units=500, activation="relu"))  # 2nd hidden layer
```

```python
ann.add(tf.keras.layers.Dense(units=output_neurons, activation="sigmoid"))  # output layer
```

```python
#ann.summary()
```

```python
ann.compile(optimizer = 'adam', loss = 'mse', metrics = ['accuracy'])
#ann.summary()
```

```python
ann.fit(X_train_norm, Y_train, batch_size=8, epochs=50)
```

```python
y_pred = ann.predict(X_test_norm)
```

```python
actual_mean=pd.DataFrame(Y_test.mean(axis=0))
actual_mean
```

```python
pred_mean = pd.DataFrame(y_pred.mean(axis=0))
pred_mean
```

```python
print('MSE (actual vs Predicted) =', mean_squared_error(Y_test,y_pred))
print('R^2 value (actual_mean vs pred_mean) = ', r2_score(actual_mean, pred_mean))
```

```python
import matplotlib.pyplot as plt

actual = actual_mean.values.flatten()
pred = pred_mean.values.flatten()

s1 = pd.Series(actual)
s2 = pd.Series(pred)

plt.figure(figsize=(20,10))
ax = plt.subplot(111)
plt.title('Actual Mean vs Pred Mean CNVs across all samples using ANN')
plt.xlabel('No of features i.e genes')
plt.ylabel('Avg CNVs across samples')
ax.plot(s1, 'b--', label='Actual')
ax.plot(s2, 'r--', label='Predicted')
ax.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.scatter(actual_mean, pred_mean)
plt.title('Corr b/w actual and predicted mean CNVs across all samples')
plt.xlabel('Actual mean CNVs across all samples')
plt.ylabel('Predicted mean CNVs across all samples')
plt.grid(True)
plt.show()
```
